```python
import nltk
from nltk.corpus import movie_reviews
import random
```

```python
documents = []
for category in movie_reviews.categories():
    for fieldId in movie_reviews.fileids(category):
        review_word_list = list(movie_reviews.words(fieldId))

        document = (review_word_list, category)
        documents.append(document)

print(documents[0])
```

```python
random.shuffle(documents)
```

```python
all_words = list(map(lambda x: x.lower(), movie_reviews.words()))
```

```python
from nltk.corpus import stopwords
english_words = stopwords.words("english")
```

```python
all_words = [word for word in all_words if len(word) > 2 and word not in english_words]
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
all_words = list(map(lambda x: lemmatizer.lemmatize(x), all_words))
```

```python
len(all_words)
```

```python
all_words = nltk.FreqDist(all_words)
print(all_words.most_common(10))
```

```python
word_features = [common_word[0] for common_word in all_words.most_common(3000)]
word_features
```

```python
def find_features(feature_doc):
    words = set(feature_doc)
    features = {}
    for word in word_features:
        is_feature_in_words = word in words
        features[word] = is_feature_in_words
    return features
```

```python
featuresets = []
for (review, category) in documents:
    feature = (find_features(review), category)
    featuresets.append(feature)
```

```python
pos_set = [t for t in featuresets if t[1] == 'pos']
neg_set = [t for t in featuresets if t[1] != 'pos']
```

```python
training_set = pos_set[:900] + neg_set[:900]
test_set = pos_set[900:] + neg_set[900:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
print("Accuracy: ", nltk.classify.accuracy(classifier, test_set))
```
